**Boxplot Structure:**

**Central Box:** The box represents the interquartile range (IQR), which contains the middle 50% of the data (from the 25th percentile, Q1, to the 75th percentile, Q3).

**Horizontal Line in the Box:** The line inside the box is the median of the dataset, which represents the 50th percentile.

**Whiskers:** These lines extend from the box to the highest and lowest values within 1.5 * IQR from the Q1 and Q3, respectively. They represent the range for the bulk of the data.

**Outliers:** Data points that lie beyond the whiskers are considered outliers. They are typically plotted as individual points.

**Multivariate method**

**Data Preprocessing:**

The dataset is divided into numerical and categorical features based on their data types.

A preprocessing pipeline is set up for both types of features:

**For numerical columns:**
Missing values are imputed with the mean of the column using SimpleImputer.
Data is scaled to have a mean of 0 and a standard deviation of 1 using **StandardScaler.**

**For categorical columns:**
Missing values are imputed with the most frequent value of the column using **SimpleImputer.**

Categories are encoded using one-hot encoding with **OneHotEncoder**, which converts categorical variable(s) into a form that could be provided to machine learning algorithms to do a better job in prediction.

**ColumnTransformer** combines these pipelines and applies the appropriate transformations to each column type.

**Dimensionality** **Reduction** **with** **PCA**:

After preprocessing, the data is converted to a dense array (if it was a sparse matrix after one-hot encoding) to apply PCA.

PCA (Principal Component Analysis) is a dimensionality reduction technique that transforms the data into a lower-dimensional space, in this case, 2 dimensions (n_components=2).

This step is often used to simplify the dataset, reduce noise, and visualize high-dimensional data in a 2-dimensional space.

**Outlier Detection with Isolation Forest:**

An **Isolation Forest** is an unsupervised learning algorithm for identifying outliers.

The IsolationForest estimator:
Fits on the preprocessed and dimensionally reduced data.

**n_estimators=100** defines the number of base estimators in the ensemble.

**contamination=0.05 **indicates the proportion of outliers in the data set.
This parameter is crucial as it affects the threshold for deciding whether a data point is an outlier.

**random_state=42** ensures reproducibility of the results by seeding the random number generator.

The model predicts the outliers in the dataset, marking them with -1 (outlier) or 1 (inlier).

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Identify numerical and categorical columns
numerical_cols = df_1.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df_1.select_dtypes(include=['object', 'category']).columns

# Define preprocessing for numerical columns (impute missing values, then scale)
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

# Define preprocessing for categorical columns (impute missing values, then OneHotEncode)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Create the preprocessing engine
preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical_cols),
    ('cat', categorical_transformer, categorical_cols)])

# Apply preprocessing
X_preprocessed = preprocessor.fit_transform(df_1)
```

```python
from sklearn.decomposition import PCA
# Instantiate PCA(Principal Component Analysis) and reduce the dimensionality of the data to two components for visualization purposes
pca = PCA(n_components=2)
# Convert the sparse matrix to a dense matrix
X_preprocessed_dense = X_preprocessed.toarray()
# Fit and transform the data using PCA
X_pca = pca.fit_transform(X_preprocessed_dense)

```

```python
from sklearn.ensemble import IsolationForest
# Instantiate the Isolation Forest algorithm for outlier detection
iso_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
 # Fit to the data and predict outliers
outliers = iso_forest.fit_predict(X_preprocessed_dense)
```

```python
import matplotlib.pyplot as plt

# Visualize the reduced data with outliers marked
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=outliers, cmap='coolwarm', edgecolor='k', s=20)
plt.title('PCA of Dataset with Outlier Indication')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

```

```python
# Remove outliers from the original preprocessed dataset
X_clean = X_preprocessed[outliers == 1]
print(f"Original preprocessed dataset shape: {X_preprocessed.shape}")
print(f"Cleaned dataset shape: {X_clean.shape}")
```

**Significant Events or Changes:** Outliers may reflect significant operational changes, environmental incidents, or regulatory changes affecting certain facilities. For instance, a sudden spike in pollutant emissions could be due to an operational mishap or a change in production volume.


**Aggregation of Diverse Entities:** If your dataset aggregates data from various types of entities (e.g., small vs. large facilities, different industries), outliers may simply represent the upper or lower bounds of this diversity. Large industrial plants might have much higher emissions compared to smaller facilities, naturally creating outliers.


**Temporal Factors:** For datasets with a time component, outliers can indicate seasonal variations, economic cycles, or specific events impacting the data. An unusually high or low reporting year might correlate with economic downturns, natural disasters, or changes in environmental regulations.


**Technical or Measurement Errors:** Inconsistencies in how data is measured, recorded, or processed can lead to outliers. This might include errors in conversion units, misplaced decimal points, or malfunctioning monitoring equipment.


**Genuine Anomalies:** Some outliers might not be errors but genuine anomalies indicating unusual but important events. These could include accidental releases, significant operational changes, or adoption of new technologies leading to drastic changes in reported figures.
